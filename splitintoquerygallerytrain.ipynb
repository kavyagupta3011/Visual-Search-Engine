```python
import os
import shutil
from tqdm import tqdm

# 1. CONFIG
partition_file = '/kaggle/input/datasets/nainika0305/deepfashionretrieval/DeepFashionRetrieval/list_eval_partition.txt'
base_img_dir = '/kaggle/input/datasets/nainika0305/deepfashionretrieval/DeepFashionRetrieval/img/img'
output_dir = '/kaggle/working/PleaseBeCorrect'

# 2. CREATE SPLIT FOLDERS
splits = ['train', 'query', 'gallery']
for split in splits:
    os.makedirs(os.path.join(output_dir, split), exist_ok=True)

# 3. READ FILE 
print("Reading partition file...")

entries = []
with open(partition_file, 'r') as f:
    lines = f.readlines()[2:] 

for line in lines:
    parts = line.strip().split()
    if len(parts) == 3:
        img_path, item_id, split = parts
        entries.append((img_path, item_id, split))

print("Total entries:", len(entries))

# 4. COPY WITH ZERO DATA LOSS
missing = 0
copied = 0

print("Copying files...")

for img_path, item_id, split in tqdm(entries):
    if img_path.startswith("img/"):
        src = os.path.join(base_img_dir, img_path.replace("img/", ""))
    else:
        src = os.path.join(base_img_dir, img_path)
    
    if not os.path.exists(src):
        print("MISSING:", src)
        missing += 1
        continue
    
    # Destination folder
    dest_folder = os.path.join(output_dir, split, item_id)
    os.makedirs(dest_folder, exist_ok=True)
    
    # UNIQUE filename
    file_name = img_path.replace("img/", "").replace("/", "_")
    
    dest = os.path.join(dest_folder, file_name)
    
    shutil.copy2(src, dest)
    copied += 1


# 5. FINAL REPORT
print("\nDone.")
print("Copied:", copied)
print("Missing:", missing)

# 6. VERIFY COUNTS
def count_images(folder):
    total = 0
    for root, _, files in os.walk(folder):
        total += len([f for f in files if f.endswith(".jpg")])
    return total

print("\nFinal split counts:")
for split in splits:
    path = os.path.join(output_dir, split)
    print(split, count_images(path))
```
